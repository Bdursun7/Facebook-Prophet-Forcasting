```python
df.replace(np.nan, inplace = True)
df_interpolated = df.interpolate()
df_interpolated
```

```python
# Veri setindeki sütunların dağılımlarına bakın
for column in df_interpolated.columns:
    sns.histplot(df_interpolated[column], kde=True)
    plt.xlabel(column)
    plt.ylabel("TEMP")
    plt.title(f"{column} Sütununun Dağılımı")
    plt.show()
```

```python
# ---- Data Transformations ----
months_dict = {'JAN':1,'FEB':2,'MAR':3,'APR':4,'MAY':5,'JUN':6,
               'JUL':7,'AUG':8,'SEP':9,'OCT':10,'NOV':11,'DEC':12}

df_interpolated = df_interpolated[['YEAR'] + list(months_dict.keys())]
```

```python
df_interpolated
```

```python
# ---melting---
df_interpolated = pd.melt(df_interpolated, 
        id_vars='YEAR',
        value_vars=months_dict.keys(),
        var_name='MONTH', 
        value_name='Sum of Value').replace({"MONTH": months_dict}) # replaces month names to month numbers.
```

```python
df_interpolated
```

```python
# ---For datetime index---
df_interpolated["DAY"] = 1
df_interpolated['DATE'] = pd.to_datetime(df_interpolated[['YEAR','MONTH','DAY']])
df_interpolated = df_interpolated.rename(columns = {'Sum of Value':'TEMP'})
df_interpolated = df_interpolated.sort_values("DATE").reset_index(drop=True)
```

```python
df_interpolated
```

```python
df_interpolated
```

```python
df_interpolated=df_interpolated.sort_values("DATE").reset_index(drop=True)
```

```python
df_interpolated
```

```python
df_interpolated = df_interpolated.drop(["YEAR", "MONTH", "DAY"], axis = 1)
```

```python
df_interpolated.dtypes
```

```python
train_size = int(len(df_interpolated) * 0.8)  # %80'i train, %20'si test verisi olarak kullanalım
train_df = df_interpolated[:train_size]
test_df = df_interpolated[train_size:]
train_df = train_df.rename(columns={'DATE': 'ds', 'TEMP': 'y'})
test_df = test_df.rename(columns={'DATE': 'ds', 'TEMP': 'y'})
```

```python
# Modeli oluşturma
model = Prophet()

# Fittetme
model.fit(train_df)
```

```python
# Test verisi üzerinde tahmin yapma
forecast = model.predict(test_df)
```

```python
# Tahmin sonuçlarını görselleştirme
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(test_df['ds'], test_df['y'], label='Gerçek Veri')  # Gerçek verileri çizdirme
ax.plot(forecast['ds'], forecast['yhat'], label='Tahmin', color='red')  # Tahminleri çizdirme
ax.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='pink', alpha=0.2)  # Belirsizlik aralığını çizdirme
plt.legend()
plt.xlabel('Tarih')
plt.ylabel('Sıcaklık')
plt.title('Sıcaklık Tahmini')
plt.show()
```

```python
from sklearn.metrics import r2_score

# R-kare değerini hesaplama
r2 = r2_score(test_df['y'], forecast['yhat'])

print("R-kare değeri:", r2)
```
