```python
# ---- Data Transformations ----
months_dict = {'JAN':1,'FEB':2,'MAR':3,'APR':4,'MAY':5,'JUN':6,
               'JUL':7,'AUG':8,'SEP':9,'OCT':10,'NOV':11,'DEC':12}

df_interpolated = df_interpolated[['YEAR'] + list(months_dict.keys())]
```

```python
df_interpolated
```

```python
# ---melting---
df_interpolated = pd.melt(df_interpolated, 
        id_vars='YEAR',
        value_vars=months_dict.keys(),
        var_name='MONTH', 
        value_name='Sum of Value').replace({"MONTH": months_dict}) # replaces month names to month numbers.
```

```python
df_interpolated
```

```python
# ---For datetime index---
df_interpolated["DAY"] = 1
df_interpolated['DATE'] = pd.to_datetime(df_interpolated[['YEAR','MONTH','DAY']])
df_interpolated = df_interpolated.rename(columns = {'Sum of Value':'TEMP'})
df_interpolated = df_interpolated.sort_values("DATE").reset_index(drop=True)
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Verileri ykleyin veya oluturun (rnek olarak rastgele scaklk verisi oluturuyorum)
existing_df = df_interpolated.copy()

# Yeni bir DataFrame oluturun ve "ds" ve "y" stunlarn uygun ekilde doldurun
df_interpolated1 = pd.DataFrame({
    'ds': pd.date_range(start='1973-01-01', periods=365, freq='D'),
    'y': existing_df['y'].values  # "existing_df" DataFrame'inden "y" stununu kopyalyoruz
})

# Verileri train ve test olarak bln
train_size = int(0.8 * len(df_interpolated1))
train_data = df_interpolated1.iloc[:train_size]
test_data = df_interpolated1.iloc[train_size:]

# Prophet modelini oluturun ve eitin
model = Prophet()
model.fit(train_data[['ds', 'y']])  # 'ds' ve 'y' stunlarn seerek eitimi gerekletirin

# leri dnem tahminleri iin bir zaman aral belirleyin (test verilerinin uzunluu kadar)
future = model.make_future_dataframe(periods=len(test_data))

# Tahminleri yapn
forecast = model.predict(future)

# Tahmin sonularn grselletirin
fig = model.plot(forecast)
plt.show()

# Test verileriyle tahmin sonularn karlatrn
fig = plt.figure(figsize=(10, 6))
plt.plot(test_data['ds'], test_data['y'], label='Gerek Veri')
plt.plot(forecast['ds'], forecast['yhat'], label='Tahmin', color='red')
plt.xlabel('Tarih')
plt.ylabel('Scaklk')
plt.legend()
plt.show()

# Tahmin sonularn detayl olarak gzlemlemek iin bileenleri grselletirin
fig = model.plot_components(forecast)
plt.show()
```
